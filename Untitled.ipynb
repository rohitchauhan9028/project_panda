```python
import pandas as pd
```

```python
df = pd.read_json(r"C:\Users\Admin\OneDrive\Documents\Downloads\dataset.json")
df
```

```python


# 1. Remove extra spaces from important text columns
df['name'] = df['name'].astype(str).str.strip()
df['category'] = df['category'].astype(str).str.strip()
df['address'] = df['address'].astype(str).str.strip()

# 2. clean  rating (convert "4.2" "-4.2")
df['rating'] = (
    df['rating']
    .astype(str)
    .str.replace(" ", "")
    .str.strip()
)

# Convert to numeric

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
 

#  3. Clean reviews (convert "2k" / "(2k)" / "987 reviews" → number)
df['reviews'] = (
    df['reviews']
        .astype(str)
        .str.replace(r'[\(\)]', '', regex=True)   # remove brackets
        .str.replace('reviews', '', regex=False)  # remove the word
        .str.replace('review', '', regex=False)
        .str.replace('k', '000')                  # 2k → 2000
        .str.replace(',', '')                     # remove comma
        .str.strip()
)


# convert Number 
df['reviews'] = pd.to_numeric(df['reviews'], errors='coerce')

# 4. Clean phone numbers
df['phone'] = df['phone'].astype(str).str.strip()

# all 
#df
# show first five
df.head()
# show last five
#df.tail()
                
```
